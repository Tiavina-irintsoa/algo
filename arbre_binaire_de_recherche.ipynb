```python
class Node: 
    key:float
    def __init__(self,**kwargs):
        if kwargs:
            self.left=kwargs.get('left')
            self.right=kwargs.get('right')
            self.key=kwargs.get('key')
            self.position=kwargs.get('position')
    def show(self):
        print(f'\t{self.key}')
        if self.left is not None:
            print(f'left of {self.key}')
            self.left.show()
        if self.right is not None:
            print(f'right of {self.key}')
            self.right.show()
    def add(self,nb):
        if self.left is None and nb<=self.key:
            self.left=Node(key=nb,position='left')
            return
        if self.right is None and nb>self.key:
            self.right=Node(key=nb,position='right')
            return
        if nb <= self.key:
            self.left.add(nb)
            return
        else:
            self.right.add(nb)
    def search(self, nb):
        #si racine
        if self.key == nb:
            return {"parent": None, "result":self}
        
        elif nb < self.key:
            if self.left is not None:
                if self.left.key == nb:
                    return {"parent": self, "result":self.left} 
                else:
                    return self.left.search(nb)
            else:
                #n'a pas d'enfant gauche alors que le nb recherhe est inferieur a lui
                return None
        else:
            if self.right is not None:
                if self.right.key == nb:
                    return {"parent": self, "result":self.right} 
                else:
                    return self.right.search(nb)
            else:
                #n'a pas d'enfant droit alors que le nb recherhe est superieur a lui
                return None
    def max(self):
        node=self
        while node.right is not None:
            node=node.right
        return node
    def afficheASC(self):
        if self.left is not None:
            self.left.afficheASC()
        print(self.key)
        if self.right is not None:
            self.right.afficheASC()
    def afficheDESC(self):
        if self.right is not None:
            self.right.afficheDESC()
        print(self.key)
        if self.left is not None:
            self.left.afficheDESC()
    def min(self):
        node=self
        while node.left is not None:
            node=node.left
        return node
```

```python
class Arbre:
    def __init__(self,**kwargs):
        self.root=kwargs.get('root')
    def construct(self,list):
        for element in list:
            self.add(element)
    def add(self,element):
        if self.root is None:
            self.root=Node(key=element)
            return 
        self.root.add(element)
    def show(self):
        self.root.show()
    def min(self):
        return self.root.min()
    def max(self):
        return self.root.max()
    def search(self,nb):
        return self.root.search(nb)
    def sortASC(self):
        self.root.afficheASC()
    def sortDESC(self):
        self.root.afficheDESC()
    def delete(self,nb):
        search_result=self.search(nb)
        #le nb n'existe pas
        if search_result is None:
            raise Exception("Number provided does not exist")
        todelete=search_result["result"]
        parent=search_result["parent"]
        
        
        if (todelete.left is None and todelete.right is None):
            print('No child')
            if parent is not None:
                print('has parent')
                if todelete.position=='right':
                    parent.right=None
                else:
                    parent.left=None
            else:
                self.root=None

            todelete=None
            
        #si a un enfant droit
        elif todelete.right is not None and todelete.left is None:
            print('has a right child')
            if parent is not None:
                print('has a parent')
                if todelete.position == 'right':
                    parent.right=todelete.right   
                else: 
                    parent.left=todelete.right
            else:
                print('has no parent')
                self.root=todelete.right
            todelete=None
        #si a un enfant gauche   
        elif todelete.left is not None and todelete.right is None:
            print('has a left child')
            if parent is not None:
                print('has no parent')
                if parent is not None:
                    if todelete.position == 'right':
                        parent.right=todelete.left
                    else:     
                        parent.left=todelete.left
            else:
                print('has no parent')
                self.root=todelete.left
            todelete=None
        else:
            print('has two children')
            max=todelete.left.max()
            if parent is not None:
                print('has  parent')
                if todelete.position == 'right':
                    parent.right=max
                else:
                    parent.left=max
                if max.left is not None:
                    search_max=self.search(max.key)
                    parent_max = search_max["parent"]
                    parent_max.right=max.left
            else:
                print('has no parent')
                self.root=max  
            max.right=todelete.right
            max=None          
```
